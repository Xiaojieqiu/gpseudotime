```julia
n = size(X)[1]

n_iter = 10000
burn = n_iter / 2

# pseudotime parameters
t = t_gt # rand(Uniform(.4999, .5001), n)
tvar = .3e-3

# kernel parameters
lambda = 1
sigma = .1

lvar = .5e-5
svar = .5e-10

mh = B_GPLVM_MH(X, n_iter, burn, t, tvar, [lambda, sigma], [lvar, svar], 1, true)
```

```julia
## Plot thinned pseudotime traces

nchoose = 50
thin = 1000
df = convert(DataFrame, mh["tchain"][:, 1:nchoose])
df[:iter] = 1:(size(df)[1])
df_thin = zeros((int(floor(size(df)[1]) / thin), size(df)[2]))

for i in 1:(size(df_thin)[1])
    j = int(thin * (i - 1) + 1)
    df_thin[i,:] = convert(DataArray, df[j,:])
end
df_thin = convert(DataFrame, df_thin)
df_melted = stack(df_thin, [1:nchoose])
names!(df_melted, [symbol(x) for x in ["variable", "value", "iter"]])

Gadfly.plot(df_melted, x = "iter", y = "value", color = "variable", Geom.line) 
```

```julia
df = convert(DataFrame, mh["theta_chain"])
#df = df[burn:end,:]
names!(df, [symbol(x) for x in ["lambda", "sigma"]])
df[:iter] = 1:(size(df)[1]) # (burn + 2)
df_melted = stack(df, [1:2]);
Gadfly.plot(df, x = "iter", y = "lambda", Geom.line)
```

```julia
Gadfly.plot(df, x = "iter", y = "sigma", Geom.line)
```

```julia
Gadfly.plot(x = t_gt, y = mean(mh["tchain"][burn:end,:], 1))
```

```julia
tp = linspace(0, 1, 50)
(lambda_map, sigma_map) = mean(mh["theta_chain"][burn:end,:], 1)
t_map = mean(mh["tchain"][burn:end,:], 1)
mu_p = predict(tp, t_map, lambda_map, sigma_map, X)

Gadfly.plot(layer(x = X[:,1], y = X[:,2], color = t_gt, Geom.point) ,
layer(x = mu_p[:,1], y = mu_p[:,2], Geom.line(preserve_order = 1), 
Theme(default_color=color("red"))))
```
